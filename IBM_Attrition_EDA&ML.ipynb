### 2. Age

```python
df.Age.describe()
```

```python
sns.boxplot(y=df.Age,color='c')
plt.show()
```

```python
sns.distplot(df[df['Attrition']==0]['Age'],color='g',label='Not Resigned')
sns.distplot(df[df['Attrition']==1]['Age'],color='r',label='Resigned')
plt.legend()
plt.show()
```

Insights:

Insights:

### Feature Engineering

```python
df_wfe=df.copy()
```

```python
df_wfe.info()
```

```python
df_wfe.tail(2)
```

```python
df_wfe['TotalSatisfaction_mean'] = (df_wfe['RelationshipSatisfaction']  + 
                                  df_wfe['EnvironmentSatisfaction'] + df_wfe['JobSatisfaction'] + 
                                  df_wfe['JobInvolvement'] + df_wfe['WorkLifeBalance'])/5
```

```python
df_wfe.tail(2)
```

```python
def NotSatif(df_wfe): 
    if  df_wfe['TotalSatisfaction_mean'] < 2.35 :
        return 1
    else: 
        return 0
df_wfe['NotSatif'] = df_wfe.apply(lambda df_wfe:NotSatif(df_wfe) ,axis = 1)
```

```python
df_wfe.tail(2)
```

```python
def SalesDpt(df_wfe):
    if df_wfe['Department']=='Sales':
        return 1
    else:
        return 0
df_wfe['SalesDpt']=df_wfe.apply(lambda df_wfe:SalesDpt(df_wfe),axis=1)
```

```python
def LowerTraining(df_wfe):
    if df_wfe['TrainingTimesLastYear']<=3:
        return 1
    else:
        return 0
df_wfe['LowerTraining']=df_wfe.apply(lambda df_wfe:LowerTraining(df_wfe),axis=1)
```

```python
def MoovingPeople(df_wfe):
    if df_wfe['NumCompaniesWorked']>4:
        return 1
    else:
        return 0
df_wfe['MoovingPeople']=df_wfe.apply(lambda df_wfe:MoovingPeople(df_wfe),axis=1)
```

```python
def LongDistance(df_wfe): 
    if  df_wfe['DistanceFromHome']>11:
        return 1
    else: 
        return 0
df_wfe['LongDistance']=df_wfe.apply(lambda df_wfe:LongDistance(df_wfe),axis=1)
```

```python
names=[]
for name,model in models:
    model.fit(X_trains,y_train)
    y_train_pred=model.predict(X_trains)
    y_train_prob=model.predict_proba(X_trains)[:,1]
    y_test_pred=model.predict(X_tests)
    y_test_prob=model.predict_proba(X_tests)[:,1]
    print(name)
    print('Performance Metrics for Train Set')
    print('Confusion Matrix - Train:','\n',confusion_matrix(y_train,y_train_pred))
    print('Accuracy Score - Train: ', accuracy_score(y_train,y_train_pred))
    print('AUC Score - Train: ', roc_auc_score(y_train,y_train_prob))
    print('Performance Metrics for Test Set')
    print('Confusion Matrix - Test:', '\n',confusion_matrix(y_test,y_test_pred))
    print('Accuracy Score - Test: ',accuracy_score(y_test,y_test_pred))
    print('AUC Score - Test: ',roc_auc_score(y_test,y_test_prob))
    print('\n')
```

```python
results=[]
names=[]
for name,model in models:
    model.fit(X_trains,y_train)
    kfold=KFold(shuffle=True,n_splits=10,random_state=0)
    cv_results=cross_val_score(model,X_trains,y_train,cv=kfold,scoring='roc_auc')
    names.append(name)
    print("%s: %f (%f)"%(name,np.mean(cv_results)*100
                         ,np.var(cv_results,ddof=1)))
    results.append((cv_results)*100)
```
